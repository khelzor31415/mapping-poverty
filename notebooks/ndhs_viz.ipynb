```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import geopandas as gp
import seaborn as sns
```

```python
ndhs = pd.read_csv('../data/ndhs/dhs_indicators.csv')
```

```python
ndhs.columns
```

```python
ndhs['wealth_index'].mean()
```

```python
TM_pal_categorical_3 = ("#ef4631", "#10b9ce", "#ff9138")
sns.set(
    style="white",
    font_scale=1.25,
    palette=TM_pal_categorical_3,
)
```

```python
ndhs['wealth_index'].hist(figsize=(15, 7.5), bins=15, density=True)
plt.title('Wealth Index Histogram', style='italic', size=20)
plt.grid(False)
sns.despine(bottom=True, left=True)
plt.tight_layout()

plt.savefig('../images/ph_wealthindex_hist.png', transparent=True, pad_inches=0, dpi=300)
```
